```python
from thornpy.geometry import get_arc_line_points, get_line_ori
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = 1
x_0 = 0
y_0 = 0
x_1 = .5
y_1 = 3
x_prev = -5
y_prev = -.1
```

```python
x_arcline, y_arcline, x_c, y_c, x_tan, y_tan, _, start_angle, i_center = get_arc_line_points(x_0, y_0, x_1, y_1, x_prev, y_prev, r)

t = np.linspace(0, 10)

plt.plot(x_c + r*np.cos(t), y_c + r*np.sin(t))

plt.scatter(x_0, y_0, c='g')
plt.scatter(x_1, y_1, c='r')
plt.plot([x_prev, x_0], [y_prev, y_0], 'b')
plt.plot(x_arcline, y_arcline, 'b')
plt.scatter(x_prev, y_prev, c='gray')
plt.scatter(x_tan, y_tan, c='b')
plt.scatter(x_c, y_c, c='k')
plt.title('{:.2f}'.format(start_angle*180/3.14))
_ = plt.axis('square')
plt.grid()
```
